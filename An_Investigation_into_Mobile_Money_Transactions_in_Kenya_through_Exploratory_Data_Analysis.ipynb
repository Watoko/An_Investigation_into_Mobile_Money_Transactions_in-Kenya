# **Mobile Money Transactions**
Mobile money transactions refer to the process of conducting payments for goods and services, as well as transferring funds to acquaintances, using portable electronic devices such as smartphones and tablets. These services are typically provided by mobile network operators or third-party entities authorized by banks.

Individuals can establish a money wallet via their mobile phones, serving as a repository for monetary value that can be utilized for purchases and transactions.

Most mobile money services enable users to send and receive funds to and from peers, store money in a virtual wallet on their mobile devices, or carry out transactions through licensed agents for withdrawal and deposit purposes.

According to research conducted by Statista, mobile wallets like PayPal, Apple Pay, and Alipay are among the most prevalent methods of mobile payments worldwide. In the Asia-Pacific region alone, there are approximately 2.8 billion active mobile wallets in use.

# Advantages of Mobile Money Transactions

Mobile money transactions offer several advantages, including:

-Providing a convenient means of sending money and conducting purchases.

-Offering a quicker payment method, with transactions processed in real-time.

-Cost-effectiveness for merchants, who can avoid expenses associated with traditional point-of-sale equipment, and reduced transaction costs for customers compared to credit card usage.

-Accessibility for all individuals with a mobile phone and necessary identification documents, such as a national identity card, to possess a digital wallet.

-Ensuring security in payments, as physical cash is susceptible to loss, while customer data is encrypted during transactions for enhanced protection.

#Mobile Money Transactions in Kenya

Mobile money transactions have become as integral to Kenyan life as a cold Tusker beer is to a Kenyan football enthusiast. The roots of mobile payments in Kenya can be traced back to 2007 when Safaricom, the country's leading telecommunications network, introduced M-Pesa.

Initially conceived to facilitate the transfer of funds to relatives and friends in rural areas, M-Pesa has since matured over its 15-year journey, resulting in:

Over 604,000 active agents spread across the Democratic Republic of Congo (DRC), Egypt, Ghana, Kenya, Lesotho, Mozambique, and Tanzania.
A customer base exceeding 51 million across seven African nations.
A staggering annual transaction volume surpassing $314 billion.

# Characteristics of Mobile Money Transactions in Kenya

Key attributes of mobile money transactions in Kenya include:

1.Interoperability: Users can seamlessly send and receive money across different mobile networks in real-time, thanks to the full interoperability of mobile money services mandated by the Central Bank of Kenya in September 2022.

2.International Remittances: Mobile money customers can engage in cross-border money transfers through various international partners like WorldRemit, Western Union, MoneyGram, and PayPal.

3.Financial Inclusion: Mobile money has significantly expanded financial access for millions of Kenyans without bank accounts or with limited banking access, offering services such as savings, loans, insurance, and investments via mobile devices.

4.Government Payments: Mobile money platforms facilitate the payment of taxes, fees, fines, and other government services through portals like KRA iTax, eCitizen, and Huduma Number.

5.Innovation and Economic Growth: Mobile money has fueled innovation and economic advancement across sectors like agriculture, healthcare, education, and e-commerce by providing a secure, affordable, and efficient payment method for goods and services.

# Dataset Overview
The dataset provides a comprehensive portrayal of monthly data pertaining to the number of active agents, registered mobile money accounts, and agent transactions involving cash inflows and outflows in Kenya spanning from January 2007 to August 2023.

# Data Source

The data originates from the Central Bank of Kenya (CBK), the regulatory authority overseeing mobile payments within Kenya. It is publicly accessible via the provided link, allowing for exportation to Excel or CSV formats, as well as conversion to PDF files.

#Data Reliability

The data is meticulously collected and published by the CBK, ensuring accuracy and reliability.

# Data Format

For the purpose of this analysis, the data is furnished in CSV file format.

# Data Attributes

The dataset encompasses the following columns:

**Year**: Denotes the fiscal year during which mobile money transactions transpire in Kenya.

**Month**: Provides monthly data regarding mobile money transactions.

**Total registered mobile money accounts**: Indicates the count of users possessing registered mobile money accounts.

**Total agent cash in cash out volume**: Represents the volume of mobile money transactions recorded by mobile money agents.
These agents, typically small mobile phone stores or retail establishments, facilitate in-person deposits and withdrawals.
"Cash in" denotes deposits, while "cash out" signifies cash withdrawals.

**Total cash in cash out value**: Reflects the monetary value of all mobile money transactions conducted in Kenya from January 2007 to February 2024.

```python
payments_df = payments.copy()

payments_df.head()
```

```python
# Create date column by Merging Year and Month column

payments_df['Date'] = pd.to_datetime(payments_df['Year'].astype(str) + '-' + payments_df['Month'])

payments_df.head()
```

```python
#Drop Year and Month Columns from payments_df

payments_df = payments_df.drop(["Year", "Month"], axis=1)
payments_df
```

```python
# Rename the following three columns:

#Total Registered Mobile Money Accounts (Millions),Total Agent Cash in Cash Out (Volume Million)
# and Total Agent Cash in Cash Out (Value KSh billions)

payments_df = payments_df.rename(columns={'Total Registered Mobile Money Accounts (Millions)': 'Mobile_Money_Accounts(Millions)', 'Total Agent Cash in Cash Out (Volume Million)': 'Cash_Deposits_Withdrawal_Volume(Million)', 'Total Agent Cash in Cash Out (Value KSh billions)': 'Cash_Deposits_Withdrawal_Value(KES billions)'})

payments_df.info()
```

```python
#Rename Active Agents column to a pythonic format

payments_df = payments_df.rename(columns={'Active Agents': 'Active_Agents'})
payments_df.head()
```

```python
#save the newly preprocessed dataset to csv format
payments_df.to_csv('payments_data.csv',index=False)
```
